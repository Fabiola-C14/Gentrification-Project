```python
# Import Dependencies
import numpy as np
import pandas as pd
from scipy import stats
import pathlib
from matplotlib import pyplot as plt
import gmaps
from configure import gkey


gmaps.configure(api_key=gkey)
```

```python
#Load Output income data to generate graphs
csv_path = pathlib.Path("Output_income_data.csv")

income_city_df = pd.read_csv(csv_path)

income_city_df.dropna()
```

```python
#List neighborhoods by income in the lower quartile
lower_percentile=income_city_df[income_city_df["Household Income"]<income_city_df["Household Income"].quantile(0.25)] 

lower_percentile
```

```python
#List neighborhoods by income in the 2rd quartile
low_middle_percentile=income_city_df[(income_city_df["Household Income"]>income_city_df["Household Income"].quantile(0.25)) & (income_city_df["Household Income"]<income_city_df["Household Income"].quantile(0.5))]
low_middle_percentile
```

```python
#List neibhorhoods by income in the 3rd qurtile
upper_middle_percentile=income_city_df[(income_city_df["Household Income"]>income_city_df["Household Income"].quantile(0.50)) & (income_city_df["Household Income"]<income_city_df["Household Income"].quantile(0.75))] 
upper_middle_percentile
```

```python
#List neighborhoods by income the upper quartile
upper_percentile=income_city_df[income_city_df["Household Income"]>income_city_df["Household Income"].quantile(0.75)] 
upper_percentile
```

```python
#Display in gmaps the income distribution in Los Angeles County
low_locations = lower_percentile[["Latitude", "Longitude"]]
low_middle_locations=low_middle_percentile[["Latitude", "Longitude"]]
upper_middle_locations=upper_middle_percentile[["Latitude", "Longitude"]]
upper_locations=upper_percentile[["Latitude", "Longitude"]]


fig = gmaps.figure()

low_layer = gmaps.symbol_layer(low_locations,fill_color='red', stroke_color='red')
low_middle_layer=gmaps.symbol_layer(low_middle_locations,fill_color='yellow', stroke_color='yellow')
upper_middle_layer=gmaps.symbol_layer(upper_middle_locations,fill_color='blue', stroke_color='blue')
upper_layer=gmaps.symbol_layer(upper_locations,fill_color='green', stroke_color='green')

fig.add_layer(low_layer)
fig.add_layer(low_middle_layer)
fig.add_layer(upper_middle_layer)
fig.add_layer(upper_layer)
fig
```
