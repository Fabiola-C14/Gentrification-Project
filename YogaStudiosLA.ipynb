```python
import pandas as pd
import csv

yoga_data = "Resources/yoga_data.csv"
zipcodes_la = "Resources/lazips.csv"

yoga_data = pd.read_csv(yoga_data)
zipcodes_la = pd.read_csv(zipcodes_la)

# # Merge the two data sets 
# yoga_studios_la = pd.merge(yoga_data)
```

```python
yoga_studios_la = pd.merge(yoga_data, zipcodes_la, how="left", on=["ZIP Code", "ZIP Code"])
yoga_studios_la
```

```python
yoga_studios_la = yoga_studios_la.rename(columns={"Postal City 1": "City",})
yoga_studios_la = yoga_studios_la.dropna()
```

```python
yoga_studios_la = yoga_studios_la.rename(columns={"Postal City 1": "City",
                                                 "Yoga Studio": "Studio"
                                                 })
yoga_studios_la.set_index("ZIP Code")
```

```python
yoga_studios_la['ZIP Code'] = yoga_studios_la['ZIP Code'].astype(str).str.rstrip('.0')
```

```python
yoga_studios_la.to_csv("lacounty_yoga_studios.csv", index=False)
```
